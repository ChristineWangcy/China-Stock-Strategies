```python
import pandas as pd
from googlefinance.client import get_price_data, get_prices_data, get_prices_time_data
reloadData = True
```

```python
# retrieve stock data using google finance
def retrieveStockData(param, fileName):
    try:
        if reloadData:
            print('Retriving data for ticker _' + param['q'] + '...')
            data = pd.DataFrame(get_price_data(param))
            data.to_csv(fileName, index=False)
        else:
            # read the existing csv
            data = pd.read_csv(fileName)
        return data
    except:
        print('Dataset could not be loaded...')
```

```python
# download single stock
def single_stock(stock_name, stock_exchange):
    param = {
        'q': stock_name, # Stock symbol
        'i': "86400", # Interval size in seconds ("86400" = 1 day intervals)
        'x': stock_exchange, # Stock exchange symbol on which stock is traded
        'P': "5Y" # Period 
    }
    fileName = stock_name + ".csv"
    data = retrieveStockData(param, fileName)
```

```python
# download several stocks
def retrieve_stocks(params, period, filename):
    df = get_prices_data(params, period)
    stocks = pd.DataFrame(df)
    display(stocks.head())
    stocks.to_csv(filename, index=True)
```
